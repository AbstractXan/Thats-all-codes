### Make_s_curve

```python
from sklearn.datasets import make_s_curve
# sample_sizes=[100,500]
# titles=['100 Sample Points','500 Sample Points']

# for i in range(2):
#     X, y = make_s_curve(n_samples=sample_sizes[i])
#     pyplot.scatter(X[:,0],X[:,2],c=y,cmap=ListedColormap(['#FF0000', '#0000FF']))
#     pyplot.suptitle(titles[i])
#     pyplot.show()   
    
noises=[0.1,0.1,0.1,0.5]
sample_sizes=[100,500,100,100]
titles=['100 Sample Points 0.1 Noise','500 Sample Points 0.1 Noise','100 Sample Points 0.1 Noise','100 Sample Points 0.5 Noise']

for i in range(4):
    X, y = make_s_curve(n_samples=sample_sizes[i], noise=noises[i])
    pyplot.scatter(X[:,0],X[:,2],c=y,cmap=ListedColormap(['#FF0000', '#0000FF']))
    pyplot.suptitle(titles[i])
    pyplot.show()      
```

```python
X, y = make_s_curve(n_samples=500, noise=0.1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)

#TRAIN Data
pyplot.scatter(X_train[:,0],X_train[:,2],c=y_train,cmap=ListedColormap(['#FF0000', '#0000FF']))
pyplot.suptitle('Train data')
pyplot.show() 

#Test Data
pyplot.scatter(X_test[:,0],X_test[:,2],c=y_test,cmap=ListedColormap(['#FF0000', '#0000FF']))
pyplot.suptitle('Test data')
pyplot.show()

```

### Make_moons

```python
from sklearn.datasets import make_moons
# generate 2d classification dataset

noises=[0.1,0.1,0.1,0.5]
sample_sizes=[100,500,100,100]
titles=['100 Sample Points','500 Sample Points','0.1 Noise','0.5 Noise']

for i in range(4):
    X, y = make_moons(n_samples=sample_sizes[i], noise=noises[i])
    pyplot.scatter(X[:,0],X[:,1],c=y,cmap=ListedColormap(['#FF0000', '#0000FF']))
    pyplot.suptitle(titles[i])
    pyplot.show()      
```

#### Test and train split

```python
X, y = make_moons(n_samples=500, noise=0.1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)

clf = LogisticRegression(solver='lbfgs')
#clf = SVC(kernel='linear',probability=True)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)

#TRAIN Data
pyplot.scatter(X_train[:,0],X_train[:,1],c=y_train,cmap=ListedColormap(['#FF0000', '#0000FF']))
pyplot.suptitle('Train data')
pyplot.show() 

#Test Data
pyplot.scatter(X_test[:,0],X_test[:,1],c=y_test,cmap=ListedColormap(['#FF0000', '#0000FF']))
pyplot.suptitle('Test data')
pyplot.show()

h = 0.02
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]

Z = Z.reshape(xx.shape)

pyplot.scatter(X_test[:,0],X_test[:,1],c=y_test,cmap=ListedColormap(['#FF0000', '#0000FF']))
pyplot.contourf(xx, yy, Z, cmap=pyplot.cm.Spectral, alpha=.4)
pyplot.show()
```

### Make_circles

```python
from sklearn.datasets import make_circles
# generate 2d classification dataset

noises=[0.1,0.1,0.1,0.5]
sample_sizes=[100,500,100,100]
titles=['100 Sample Points','500 Sample Points','0.1 Noise','0.5 Noise']

for i in range(4):
    X, y = make_circles(n_samples=sample_sizes[i], noise=noises[i])
    pyplot.scatter(X[:,0],X[:,1],c=y,cmap=ListedColormap(['#FF0000', '#0000FF']))
    pyplot.suptitle(titles[i])
    pyplot.show()          
```

```python
X, y = make_circles(n_samples=100, noise=0.1)
test_sizes=[0.4,0.6]
titles=[' :0.4 Test Size' , ' :0.6 Test Size']
for i in range(2):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_sizes[i], random_state=2)
    clf = LogisticRegression(solver='lbfgs')
    clf.fit(X_train,y_train)
    y_pred = clf.predict(X_test)

    #TRAIN Data
    pyplot.scatter(X_train[:,0],X_train[:,1],c=y_train,cmap=ListedColormap(['#FF0000', '#0000FF']))
    pyplot.suptitle('Train data'+titles[i])
    pyplot.show() 

    #Test Data
    pyplot.scatter(X_test[:,0],X_test[:,1],c=y_test,cmap=ListedColormap(['#FF0000', '#0000FF']))
    pyplot.suptitle('Test data'+titles[i])
    pyplot.show()

    h = 0.02
    x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]

    Z = Z.reshape(xx.shape)

    pyplot.scatter(X_test[:,0],X_test[:,1],c=y_test,cmap=ListedColormap(['#FF0000', '#0000FF']))
    pyplot.contourf(xx, yy, Z, cmap=pyplot.cm.Spectral, alpha=.4)
    pyplot.suptitle('Contours'+titles[i])
    pyplot.show()
```

### Make_blobs


```python
from sklearn.datasets import make_blobs

# generate 2d classification dataset
sample_sizes=[100,500]
titles=['100 Sample Points','500 Sample Points']

for i in range(2):
    X, y = make_blobs(n_samples=sample_sizes[i],centers=2, n_features=2)
    # scatter plot, dots colored by class value
    pyplot.scatter(X[:,0],X[:,1],c=y,cmap=ListedColormap(['#FF0000', '#0000FF']))
    pyplot.suptitle(titles[i])
    pyplot.show() 
```

```python
X, y = make_blobs(n_samples=100,centers=2, n_features=2)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)

clf = LogisticRegression(solver='lbfgs')
#clf = SVC(kernel='linear',probability=True)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)

#TRAIN Data
pyplot.scatter(X_train[:,0],X_train[:,1],c=y_train,cmap=ListedColormap(['#FF0000', '#0000FF']))
pyplot.suptitle('Train data')
pyplot.show() 

#Test Data
pyplot.scatter(X_test[:,0],X_test[:,1],c=y_test,cmap=ListedColormap(['#FF0000', '#0000FF']))
pyplot.suptitle('Test data')
pyplot.show()

h = 0.02
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]

Z = Z.reshape(xx.shape)

pyplot.scatter(X_test[:,0],X_test[:,1],c=y_test,cmap=ListedColormap(['#FF0000', '#0000FF']))
pyplot.contourf(xx, yy, Z, cmap=pyplot.cm.Spectral, alpha=.4)
pyplot.show()
```
