```python
from random import *

#Generate random string of size 'size' and string of zeroes of size 'size'
def generate(size,rand):
    temp=[]
    if rand==0:
        temp = [0 for i in range(size)] #Generates a list of bits of size partition_length
    else:
        temp = [randint(0,1) for i in range(size)] #Generates a random list of bits of size partition_length
    summ = ''.join(list(map(str,temp))) #Converts list of bits to string
    return summ
    
#Add function on individual bits
def add_digits(a,b):
    return bin(int(a,2)^int(b,2))[2:]

#Checks if adding a,b,c gives a carry
def check_carry(a,b,c):
    if ((a == '1' and b == '1') or (b =='1' and c == '1') or (c == '1' and a == '1')):
        return '1'
    return '0'

#Adds bitSequences together : Assuming both are of same length
def add(a,b):
    
    # summ is string of all zeroes of length partition_length
    summ = generate(len(a),0)
    
    carry='0' #Initialize carry to zero
    
    for i in range(len(a)-1,-1,-1): #Add digits from behind
        summ_bit = add_digits( add_digits(a[i],b[i]), carry)[0]
        carry=check_carry(a[i],b[i],carry)
        
        #Update summ bitSequence at ith position
        summ1 = summ[:i]
        summ1 += summ_bit
        summ = summ1 + summ[i+1:]
    
    #If carry, add it to the sumBitsequence
    if carry=='1':
        # summ is string of all zeroes of length partition_length-1
        tempsumm = generate(len(a)-1,0)
        summ , carry = add(summ,tempsumm+carry)
    
    return summ,carry

#Inverts given bitSequence
def invert(a):
    out=''
    for i in range(len(a)):
        if a[i]=='1':
            out+='0'
        else:
            out+='1'
    return out

#Sender returns a messsage with the CheckSum at the end of the randomly generated message
def sender(partition_length,no_of_partitions):
    partitions = []
    
    #Generating partitions
    for i in range(no_of_partitions):
        bitSequence = generate(partition_length,1)
        partitions.append(bitSequence)
    #print(partitions) 
    
    #Initializing sum
    summ = generate(partition_length,0)
    
    message = ''
    #Adding all the partitions
    for part in partitions:
        message += part
        summ , _ = add(summ,part)
    
    message += invert(summ)
    print("Message sent:" , message)
    return message

#Reciever function
def reciever(message,partition_length):
    partitions=[]
    temp='' 
    temp_num=0
    for i in range(len(message)):
        if(temp_num>=partition_length):
            partitions.append(temp)
            temp=''
            temp_num=0
        temp+=message[i]
        temp_num+=1
    partitions.append(temp)
    print("Message Partitions Recievd: ",partitions)
    
    #Initializing sum
    summ = generate(partition_length,0)
    
    #Adding all partitions
    for part in partitions:
        summ , _ = add(summ,part)
    print("Sum of all partitions: ",summ)
    
    
    #Validating sum to be all 1s
    for i in range(len(summ)):
        if summ[i] == '0':
            return 'Error'
    return 'Correctly recieved'

reciever(sender(16,3),16) #16 bit per partition #3 partitions
```
